```python
import pandas as pd
from pathlib import Path
import re
import itertools
```

```python
CSV_PATH = Path('data/csv')
GAMEDEV_PATH = CSV_PATH / 'gamedev.stackexchange.com'
```

```python
posts_df = pd.read_csv(GAMEDEV_PATH / 'Posts.csv')
```

```python
def get_tags(posts):
    tag_pattern = re.compile(r'<(?P<tag>[^<^>]+)>')
    for tags_str in posts.loc[posts.PostTypeId == 1, 'Tags']:
        tags = tag_pattern.findall(tags_str)
        tags = list(sorted(filter(None, tags)))
        if len(tags) > 1:
            for tag_pair in itertools.combinations(tags, 2):
                yield tag_pair

def get_tag_pairs(posts):
    return pd.DataFrame(get_tags(posts_df), columns=['tag1', 'tag2'])

def get_most_popular_tag_pairs(posts):
    tag_pairs = get_tag_pairs(posts)
    return tag_pairs.groupby(['tag1', 'tag2'])\
                    .size()\
                    .reset_index()\
                    .rename(columns={0: 'count'})\
                    .sort_values('count', ascending=False)\
                    .reset_index(drop=True)
```

```python
most_popular_tag_pair_df = get_most_popular_tag_pairs(posts_df)
most_popular_tag_pair_df.head(16)
```
